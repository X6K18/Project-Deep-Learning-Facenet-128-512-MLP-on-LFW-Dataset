```python
import os 
import shutil
```

```python
source_dir = 'D:/facenet/deep_face/deep_face/data/lfw_funneled'
```

```python
target_dir = 'D:/facenet/deep_face/deep_face/data/lfw_filtered_10'
```

```python
MIN_IMAGES = 10
```

```python
count_person = 0

for person_name in os.listdir(source_dir):
    person_path = os.path.join(source_dir, person_name)
    
    if os.path.isdir(person_path):
        images = [img for img in os.listdir(person_path) 
                  if img.lower().endswith(('.jpg', '.png', '.jpeg'))]
        
        if len(images) >= MIN_IMAGES:
            target_person_path = os.path.join(target_dir, person_name)
            shutil.copytree(person_path, target_person_path)
            
            count_person += 1
            print(f"Đã copy: {person_name} ({len(images)} ảnh)")

print(f"\nTổng số người thỏa điều kiện: {count_person}")
```
